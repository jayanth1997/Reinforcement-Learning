```python
restore_file_path = "gdrive/My Drive/Colab Notebooks/cartpole_dqn_final_20.h5"
```

```python
env = gym.make("CartPole-v0")
model = load_model(restore_file_path)
state = env.reset()
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
nte=1
en=0
frames = []

while(en < nte):
    state = env.reset()
    done=False
    state = np.reshape(state, [1, state_size])
    episode_rew = 0
    while not done:
        frames.append(env.render(mode = 'rgb_array'))
        act_values = model.predict(state)
        action=np.argmax(act_values[0])  # returns action
        observe, reward, done, info = env.step(action)
        state=np.reshape(observe, [1, state_size])
        episode_rew += reward
    en+=1
    print("Episode reward", episode_rew)
```

```python
env.render()
plt.figure(figsize=(frames[0].shape[1] / 72.0, frames[0].shape[0] / 72.0), dpi = 72)
patch = plt.imshow(frames[0])
plt.axis('off')
animate = lambda i: patch.set_data(frames[i])
ani = matplotlib.animation.FuncAnimation(plt.gcf(), animate, frames=len(frames), interval = 50)
HTML(ani.to_jshtml()) 
Writer = matplotlib.animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
ani.save('/content/gdrive/My Drive/Colab Notebooks/Cartpole_dqn_mp4.mp4', writer=writer)
```
